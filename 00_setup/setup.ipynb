```python
# Criar o catalog principal do projeto
spark.sql("""
    CREATE CATALOG IF NOT EXISTS cosmic_insights
    COMMENT 'Lakehouse de dados espaciais da NASA - Projeto Cosmic Insights'
""")

# Usar o catalog
spark.sql("USE CATALOG cosmic_insights")

# Criar schemas (databases) para cada camada
spark.sql("""
    CREATE SCHEMA IF NOT EXISTS bronze
    COMMENT 'Camada Bronze - Dados brutos da NASA'
""")

spark.sql("""
    CREATE SCHEMA IF NOT EXISTS silver
    COMMENT 'Camada Silver - Dados limpos e transformados'
""")

spark.sql("""
    CREATE SCHEMA IF NOT EXISTS gold
    COMMENT 'Camada Gold - Agregações e métricas de negócio'
""")

print("✅ Catalog 'cosmic_insights' criado com sucesso!")
print("✅ Schemas: bronze, silver, gold")
```
